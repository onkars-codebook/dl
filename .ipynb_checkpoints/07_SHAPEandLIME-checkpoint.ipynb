```python
# ============================================================
# INSTALL REQUIRED LIBRARIES
# ============================================================

!pip install shap lime scikit-image -q

# ============================================================
# IMPORT LIBRARIES
# ============================================================

import tensorflow as tf
import shap
import lime
import numpy as np
import matplotlib.pyplot as plt

from lime import lime_image
from skimage.segmentation import mark_boundaries

from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dense
```

```python
# ============================================================
# LOAD MNIST DATASET
# ============================================================

# MNIST contains handwritten digit images (0–9)

(X_train, y_train), (X_test, y_test) = mnist.load_data()

# ============================================================
# PREPROCESS DATASET
# ============================================================

# Reshape images into 28x28x1 format
# Normalize pixel values between 0 and 1

X_train = X_train.reshape(-1, 28, 28, 1) / 255.0

X_test = X_test.reshape(-1, 28, 28, 1) / 255.0

# ============================================================
# DISPLAY DATASET INFORMATION
# ============================================================

print("Training Shape :", X_train.shape)

print("Testing Shape :", X_test.shape)

print("Number of Classes :", len(np.unique(y_train)))
```

```python
# ============================================================
# DISPLAY SAMPLE DIGITS
# ============================================================

plt.figure(figsize=(10,2))

for i in range(5):

    plt.subplot(1,5,i+1)

    plt.imshow(
        X_train[i].reshape(28,28),
        cmap='gray'
    )

    plt.title(f"Digit : {y_train[i]}")

    plt.axis('off')

plt.show()
```

```python
# ============================================================
# BUILD CNN MODEL
# ============================================================

# CNN = Convolutional Neural Network
# Used mainly for image classification

model = Sequential([

    # Convolution Layer
    Conv2D(
        32,
        (3,3),
        activation='relu',
        input_shape=(28,28,1)
    ),

    # Pooling Layer
    MaxPooling2D((2,2)),

    # Convert 2D features into 1D
    Flatten(),

    # Fully Connected Layer
    Dense(
        64,
        activation='relu'
    ),

    # Output Layer (10 classes)
    Dense(
        10,
        activation='softmax'
    )
])

# ============================================================
# COMPILE MODEL
# ============================================================

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# ============================================================
# DISPLAY MODEL ARCHITECTURE
# ============================================================

model.summary()
```

```python
# ============================================================
# TRAIN CNN MODEL
# ============================================================

history = model.fit(

    X_train,
    y_train,

    epochs=2,

    validation_split=0.2,

    verbose=1
)

# ============================================================
# EVALUATE MODEL
# ============================================================

loss, accuracy = model.evaluate(
    X_test,
    y_test
)

print("\nTest Accuracy :", accuracy)
```

```python
# ============================================================
# SHAP EXPLAINABILITY
# ============================================================

# SHAP explains feature importance
# for deep learning predictions

# Create SHAP explainer

explainer = shap.DeepExplainer(
    model,
    X_train[:100]
)

# Generate SHAP values for one image

shap_values = explainer.shap_values(
    X_test[:1]
)

print("SHAP Explanation Generated Successfully")

# ============================================================
# DISPLAY SHAP IMAGE
# ============================================================

shap.image_plot(
    shap_values,
    X_test[:1]
)
```

```python
# ============================================================
# LIME EXPLAINABILITY
# ============================================================

# Create LIME image explainer

explainer_lime = lime_image.LimeImageExplainer()

# ============================================================
# CONVERT GRAYSCALE IMAGE TO RGB
# ============================================================

# LIME expects RGB image

img = np.repeat(
    X_test[0],
    3,
    axis=2
)

# ============================================================
# PREDICTION FUNCTION FOR LIME
# ============================================================

def predict_fn(images):

    # Convert RGB back to grayscale

    gray = np.array(images)[:,:,:,0].reshape(-1,28,28,1)

    return model.predict(gray)

# ============================================================
# GENERATE LIME EXPLANATION
# ============================================================

explanation = explainer_lime.explain_instance(

    img.astype('double'),

    classifier_fn=predict_fn,

    top_labels=1,

    hide_color=0,

    num_samples=1000
)

# ============================================================
# GET IMPORTANT REGIONS
# ============================================================

temp, mask = explanation.get_image_and_mask(

    explanation.top_labels[0],

    positive_only=True,

    num_features=5,

    hide_rest=False
)

# ============================================================
# DISPLAY LIME OUTPUT
# ============================================================

plt.figure(figsize=(5,5))

plt.imshow(
    mark_boundaries(temp, mask)
)

plt.title("LIME Explanation")

plt.axis('off')

plt.show()

print("LIME Explanation Generated Successfully")
```

```python
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt

# Load dataset
(X_train, _), _ = mnist.load_data()
X_train = (X_train.reshape(-1, 28, 28, 1).astype("float32") - 127.5) / 127.5

# Generator
generator = tf.keras.Sequential([
    layers.Dense(128, activation='relu', input_shape=(100,)),
    layers.Dense(28*28*1, activation='tanh'),
    layers.Reshape((28,28,1))
])

# Discriminator
discriminator = tf.keras.Sequential([
    layers.Flatten(input_shape=(28,28,1)),
    layers.Dense(128, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])

# Compile discriminator
discriminator.compile(optimizer='adam',
                      loss='binary_crossentropy')

# GAN
discriminator.trainable = False

gan = tf.keras.Sequential([generator, discriminator])

gan.compile(optimizer='adam',
            loss='binary_crossentropy')

# Training
batch_size = 64

for epoch in range(1000):

    # Train discriminator
    idx = tf.random.uniform([batch_size], 0, X_train.shape[0], dtype=tf.int32)
    real = tf.gather(X_train, idx)

    noise = tf.random.normal([batch_size, 100])
    fake = generator.predict(noise, verbose=0)

    discriminator.train_on_batch(real, tf.ones((batch_size,1)))
    discriminator.train_on_batch(fake, tf.zeros((batch_size,1)))

    # Train generator
    noise = tf.random.normal([batch_size, 100])
    gan.train_on_batch(noise, tf.ones((batch_size,1)))

# Generate image
noise = tf.random.normal([1,100])
img = generator.predict(noise, verbose=0)[0,:,:,0]

# Show generated digit
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.show()


```
