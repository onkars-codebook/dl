```python
# ============================================================
# IMPORT LIBRARIES
# ============================================================

import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import SimpleRNN
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import GRU
from tensorflow.keras.layers import Embedding
from tensorflow.keras.layers import Bidirectional
```

```python
# ============================================================
# LOAD IMDB DATASET
# ============================================================

# num_words = top 10000 most frequent words

vocab_size = 10000

(x_train, y_train), (x_test, y_test) = imdb.load_data(
    num_words=vocab_size
)

print("Training Samples :", len(x_train))
print("Testing Samples :", len(x_test))
```

```python
# ============================================================
# DISPLAY SAMPLE REVIEW
# ============================================================

print("Sample Review (Encoded):")
print(x_train[0])

print("\nSentiment Label:")
print(y_train[0])

# 0 = Negative
# 1 = Positive
```

```python
# ============================================================
# PADDING SEQUENCES
# ============================================================

# Reviews have different lengths
# So we make all reviews same length

max_length = 200

x_train = pad_sequences(
    x_train,
    maxlen=max_length
)

x_test = pad_sequences(
    x_test,
    maxlen=max_length
)

print("Shape after Padding:")
print(x_train.shape)
```

```python
# ============================================================
# SIMPLE RNN MODEL
# ============================================================

simple_rnn_model = Sequential()

# Word Embedding Layer
simple_rnn_model.add(
    Embedding(vocab_size, 128, input_length=max_length)
)

# Simple RNN Layer
simple_rnn_model.add(
    SimpleRNN(64)
)

# Output Layer
simple_rnn_model.add(
    Dense(1, activation='sigmoid')
)

simple_rnn_model.summary()
```

```python
# ============================================================
# COMPILE SIMPLE RNN MODEL
# ============================================================

simple_rnn_model.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

print("Simple RNN Model Compiled")
```

```python
# ============================================================
# TRAIN SIMPLE RNN MODEL
# ============================================================

history_rnn = simple_rnn_model.fit(
    x_train,
    y_train,
    epochs=3,
    batch_size=64,
    validation_split=0.2
)
```

```python
# ============================================================
# EVALUATE SIMPLE RNN
# ============================================================

rnn_loss, rnn_accuracy = simple_rnn_model.evaluate(
    x_test,
    y_test
)

print("Simple RNN Accuracy :", rnn_accuracy)
```

```python
# ============================================================
# LSTM MODEL
# ============================================================

lstm_model = Sequential()

lstm_model.add(
    Embedding(vocab_size, 128, input_length=max_length)
)

# LSTM Layer
lstm_model.add(
    LSTM(64)
)

lstm_model.add(
    Dense(1, activation='sigmoid')
)

lstm_model.summary()
```

```python
# ============================================================
# COMPILE LSTM MODEL
# ============================================================

lstm_model.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

print("LSTM Model Compiled")
```

```python
# ============================================================
# TRAIN LSTM MODEL
# ============================================================

history_lstm = lstm_model.fit(
    x_train,
    y_train,
    epochs=3,
    batch_size=64,
    validation_split=0.2
)
```

```python
# ============================================================
# EVALUATE LSTM MODEL
# ============================================================

lstm_loss, lstm_accuracy = lstm_model.evaluate(
    x_test,
    y_test
)

print("LSTM Accuracy :", lstm_accuracy)
```

```python
# ============================================================
# GRU MODEL
# ============================================================

gru_model = Sequential()

gru_model.add(
    Embedding(vocab_size, 128, input_length=max_length)
)

# GRU Layer
gru_model.add(
    GRU(64)
)

gru_model.add(
    Dense(1, activation='sigmoid')
)

gru_model.summary()
```

```python
# ============================================================
# COMPILE GRU MODEL
# ============================================================

gru_model.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

print("GRU Model Compiled")
```

```python
# ============================================================
# TRAIN GRU MODEL
# ============================================================

history_gru = gru_model.fit(
    x_train,
    y_train,
    epochs=3,
    batch_size=64,
    validation_split=0.2
)
```

```python
# ============================================================
# EVALUATE GRU MODEL
# ============================================================

gru_loss, gru_accuracy = gru_model.evaluate(
    x_test,
    y_test
)

print("GRU Accuracy :", gru_accuracy)
```

```python
# ============================================================
# BIDIRECTIONAL LSTM MODEL
# ============================================================

bidirectional_model = Sequential()

bidirectional_model.add(
    Embedding(vocab_size, 128, input_length=max_length)
)

# Bidirectional Layer
bidirectional_model.add(
    Bidirectional(LSTM(64))
)

bidirectional_model.add(
    Dense(1, activation='sigmoid')
)

bidirectional_model.summary()
```

```python
# ============================================================
# COMPILE BIDIRECTIONAL MODEL
# ============================================================

bidirectional_model.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

print("Bidirectional LSTM Model Compiled")
```

```python
# ============================================================
# COMPILE BIDIRECTIONAL MODEL
# ============================================================

bidirectional_model.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

print("Bidirectional LSTM Model Compiled")
```

```python
# ============================================================
# EVALUATE BIDIRECTIONAL MODEL
# ============================================================

bidirectional_loss, bidirectional_accuracy = bidirectional_model.evaluate(
    x_test,
    y_test
)

print("Bidirectional LSTM Accuracy :", bidirectional_accuracy)
```

```python
# ============================================================
# COMPARE ALL MODELS
# ============================================================

models = [
    "Simple RNN",
    "LSTM",
    "GRU",
    "Bidirectional LSTM"
]

accuracies = [
    rnn_accuracy,
    lstm_accuracy,
    gru_accuracy,
    bidirectional_accuracy
]

for model, accuracy in zip(models, accuracies):
    print(model, ":", accuracy)
```

```python
# ============================================================
# PLOT MODEL ACCURACIES
# ============================================================

plt.figure(figsize=(8,5))

plt.bar(models, accuracies)

plt.xlabel("Models")
plt.ylabel("Accuracy")
plt.title("RNN Architecture Comparison")

plt.show()
```

```python
# ============================================================
# TEST CUSTOM REVIEW
# ============================================================

word_index = imdb.get_word_index()

# Example review
review = "This movie was fantastic and amazing"

# Convert words into numbers
encoded_review = []

for word in review.lower().split():

    if word in word_index:
        encoded_review.append(word_index[word] + 3)

# Padding
encoded_review = pad_sequences(
    [encoded_review],
    maxlen=max_length
)

# Prediction
prediction = lstm_model.predict(encoded_review)

print("Prediction Value :", prediction)

if prediction > 0.5:
    print("Positive Review")
else:
    print("Negative Review")
```
